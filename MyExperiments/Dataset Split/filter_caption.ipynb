```python
import os
import json

# === CONFIG ===
pt_folder = "/home/woody/iwi5/iwi5204h/HistGen/Data/WSI/PatchResult/pt_files/dinov2_vitl"  # Replace with your .pt files directory
caption_file = "train.json"  # Replace with your original caption file
output_file = "filtered_captions.json"  # Output file with 8352 valid entries

# === Step 1: Get list of .pt filenames (without extension) ===
pt_files = [f for f in os.listdir(pt_folder) if f.endswith(".pt")]
pt_ids = set(os.path.splitext(f)[0] for f in pt_files)

print(f"Found {len(pt_ids)} .pt files.")

# === Step 2: Load the original caption file ===
with open(caption_file, "r") as f:
    captions = json.load(f)

print(f"Original caption entries: {len(captions)}")

# === Step 3: Filter captions to those with .pt files ===
filtered_captions = [entry for entry in captions if entry["id"].replace(".tiff", "") in pt_ids]

print(f"Filtered caption entries: {len(filtered_captions)}")

# === Step 4: Save the filtered list ===
with open(output_file, "w") as f:
    json.dump(filtered_captions, f, indent=2)

print(f"Filtered captions saved to {output_file}")
```
